```python
import pandas as pd
import numpy as np

#-----------------------------------------------------#
with open('patient_data.csv', 'r') as file:
  lines = file.readlines()

new_lines = []
for line in lines:                                                                                     #удаление дубликатов в данных о пациенте
  if not line in new_lines:
    new_lines.append(line)

with open('patient_data.csv', 'w') as file:
  file.write('\n'.join(new_lines))

#-----------------------------------------------------#
with open('mri_data.csv', 'r') as file:
  lines = file.readlines()

new_lines = []
for line in lines:
  if not line in new_lines:
    new_lines.append(line)                                                                  #удаление дубликатов в данных о МРТ пациента 

with open('mri_data.csv', 'w') as file:
  file.write('\n'.join(new_lines))

#-----------------------------------------------------#

df1= pd.read_csv('patient_data.csv', delimiter=';' ,  mangle_dupe_cols=True) #добавление информации о пациенте 
df2= pd.read_csv('mri_data.csv', delimiter=';',  mangle_dupe_cols=True)        #добавление МРТ пациента

#-----------------------------------------------------#

df3 = df1.merge(df2, how='left')                              #соединение в один файл пациента и его МРТ

#-----------------------------------------------------#

df3['MR Delay'] = df3['MR Delay'].fillna('0')           #замена всех пустых мест в задержке на 0

#-----------------------------------------------------#

df3['Group'] = 1                                                           #добавление новой колонки Group
df3["Group"]   = np.where(df3["CDR"] > 0, "Деменция", "Нет деменции")   #если CDR>0, то деменция, иначе ее нет

#-----------------------------------------------------#

df3["Gender"] = np.where(df3["Gender"] == "M", "Мужской", "Женский")  #Замена F,M на Женский, Мужской

#-----------------------------------------------------#


df3 = df3 [  ['Subject ID', 'MRI ID', 'Group' , 'Visit', 'MR Delay', 'Gender',  'Age',  'EDUC',  'MMSE', 'CDR',  'eTIV', 'nWBV',  'ASF']  ] #перестановка столбцов в нужном порядке

#-----------------------------------------------------#

df3['MR Delay'] = pd.to_numeric(df3['MR Delay']) / 60  #преобразование секунд в минуты в Delay
df3['MR Delay'] = df3['MR Delay'].round(2)                      # округление числа до двух чисел после запятой в Delay

#-----------------------------------------------------#

df3['MR Delay'] = df3['MR Delay'].astype(float) #преобразование из стринга в флоат 
df3['ASF'] = df3['ASF'].astype(float)                      #преобразование из стринга в флоат 

#-----------------------------------------------------#

df3.to_csv("tsvresult.tsv", sep = '\t')                   # скачивание файла в tsv формате 

#-----------------------------------------------------#


df3 = df3.rename(columns={'Subject ID': 'ID пациента', 'MRI ID': 'МРТ ID',  'Group': 'Группа', 'Visit': 'Визит №',   #замена названия таблицы на русский 
                                                                                        'MR Delay': 'МР Задержка', 'Gender': 'Пол', 'Age': 'Возраст',
                            'EDUC': 'Образование, лет', 'MMSE': 'Оценка псих. состояния', 'CDR': 'Уровень деменции',
                                                     'eTIV': 'Внутречерепной объем', 'nWBV': 'Нормализованный объем мозга', })

#-----------------------------------------------------#
 
df3.to_excel("xlresult.xlsx")                                  # скачивание файла в xlsx формате 

#-----------------------------------------------------#


```
